### Fields or Column Headers

* **Field/Column 1:** X = Longitude coordinate for squirrel sighting point (int)

* **Field/Column 2:** Y = Latitude coordinate for squirrel sighting point (int)

* **Field/Column 3:** Unique Squirrel ID = Identification tag for each squirrel sightings. The tag is comprised of "Hectare ID" + "Shift" + "Date" + "Hectare Squirrel Number." (string)

* **Field/Column 4:** Hectare = ID tag, which is derived from the hectare grid used to divide and count the park area. One axis that runs predominantly north-to-south is numerical (1-42), and the axis that runs predominantly east-to-west is roman characters (A-I). (string)

* **Field/Column 5:** Shift = Value is either "AM" or "PM," to communicate whether or not the sighting session occurred in the morning or late afternoon. (string)

* **Field/Column 6:** Date = Concatenation of the sighting session day and month. (string)

* **Field/Column 7:** Hectare Squirrel Number = Number within the chronological sequence of squirrel sightings for a discrete sighting session. (int)

* **Field/Column 8:** Age = Value is either "Adult" or "Juvenile." (string)

* **Field/Column 9:** Primary Fur Color = Value is either "Gray," "Cinnamon" or "Black." (string)

* **Field/Column 10:** Highlight Fur Color = Discrete value or string values comprised of "Gray," "Cinnamon" or "Black." (string)

* **Field/Column 11:** Combination of Primary and Highlight Color = A combination of the previous two columns; this column gives the total permutations of primary and highlight colors observed. (string)

* **Field/Column 12:** Color notes = Sighters occasionally added commentary on the squirrel fur conditions. These notes are provided here. (string)

* **Field/Column 13:** Location = Value is either "Ground Plane" or "Above Ground." Sighters were instructed to indicate the location of where the squirrel was when first sighted. (string)

* **Field/Column 14:** Above Ground Sighter Measurement = For squirrel sightings on the ground plane, fields were populated with a value of “FALSE.” (string)

* **Field/Column 15:** Specific Location = Sighters occasionally added commentary on the squirrel location. These notes are provided here. (string)

* **Field/Column 16:** Running = Squirrel was seen running. (checkbox?)

* **Field/Column 17:** Chasing = Squirrel was seen chasing another squirrel.(checkbox)

* **Field/Column 18:** Climbing = Squirrel was seen climbing a tree or other environmental landmark.(checkbox)

* **Field/Column 19:** Eating = Squirrel was seen eating. (checkbox)

* **Field/Column 20:** Foraging = Squirrel was seen foraging for food.(checkbox)

* **Field/Column 21:** Other Activities (string)

* **Field/Column 22:** Kuks = Squirrel was heard kukking, a chirpy vocal communication used for a variety of reasons. (checkbox)

* **Field/Column 23:** Quaas = Squirrel was heard quaaing, an elongated vocal communication which can indicate the presence of a ground predator such as a dog. (checkbox)

* **Field/Column 24:** Moans = Squirrel was heard moaning, a high-pitched vocal communication which can indicate the presence of an air predator such as a hawk. (checkbox)

* **Field/Column 25:** Tail flags = Squirrel was seen flagging its tail. Flagging is a whipping motion used to exaggerate squirrel's size and confuse rivals or predators. Looks as if the squirrel is scribbling with tail into the air. (checkbox)

* **Field/Column 26:** Tail twithes = Squirrel was seen twitching its tail. Looks like a wave running through the tail, like a breakdancer doing the arm wave. Often used to communicate interest, curiosity.(checkbox)

* **Field/Column 27:** Approaches = Squirrel was seen approaching human, seeking food. (checkbox)

* **Field/Column 28:** Indifferent = Squirrel was indifferent to human presence. (checkbox)

* **Field/Column 29:** Runs from = Squirrel was seen running from humans, seeing them as a threat. (checkbox)

* **Field/Column 30:** Other Interactions = Sighter notes on other types of interactions between squirrels and humans. (string)

* **Field/Column 31:** Lat/Long = Latitude and longitude (string)

##Part 3: Extract / Transform

I created lists to store the information for each of the columns that I plan to use in my analysis. Most of the data I am using is categorical, and have pre-determined categories. So, in terms of cleaning, I only included rows where the data existed/correct format and ignored rows with missing data. Additionally, I need to change some of the string types in the data to other Python types. For example, the 'Running' column has either the string 'true' or 'false', and I want to change it to boolean types of either True or False. These are the datasets that I am using in my analysis:
  * Age --> int (0 and 1)
  * Color --> String
  * Shift --> String
  * Hectare squirrel number --> int
  * Running --> boolean (normalize true to True and false to False)
  * Chasing --> boolean
  * Climbing --> boolean
  * Eating --> boolean
  * Foraging --> boolean



```python
# Create lists for each of these columns and store the information
age = [] # index 7
color = [] # index 8 
shift = [] # index 4
hectare = [] # index 6
run = [] # index 15
chase = [] # index 16
climb = [] # index 17
eat = [] # index 18
forage = [] # index 19

am_count = 0
pm_count = 0

# '/content/drive/MyDrive/Senior_Year/Fall_2021/Data_Management_and_Analysis/HW#2/2018_Central_Park_Squirrel_Census_-_Squirrel_Data.csv'
with open('data/raw/2018_Central_Park_Squirrel_Census_-_Squirrel_Data.csv') as f:
  next(f) # skip header
  reader = csv.reader(f)
  for row in reader: # row is a list of strings!
    # Step 1: age - convert 
    # for age there are two options = "Adult" or "Juvenile"
    if (row[7].lower() == 'adult' or row[7].lower() == 'juvenile'):
      age.append(row[7].lower())

    # Step 2: color
    # "Gray," "Cinnamon" or "Black."
    if (row[8].lower() == 'gray' or row[8].lower() == 'cinnamon' or row[8].lower() == 'black'):
      color.append(row[8].lower())

    # Step 3: shift
    if (row[4].upper() == 'AM' or row[4].upper() == 'PM'):
      shift.append(row[4].upper())
      if (row[4].upper() == 'AM'):
        am_count += int(row[6])
      elif (row[4].upper() == 'PM'):
        pm_count += int(row[6])

    # Step 4: hectare - convert to integer 
    if (int(row[6])):
      hectare.append(int(row[6]))
    
    # Step 5 - convert to bool
    # run = index 15
    if (row[15].lower() == 'true'):
        run.append(True)
    elif (row[15].lower() == 'false'):
        run.append(False)
    # chase = index 16
    if (row[16].lower() == 'true'):
        chase.append(True)
    elif (row[16].lower() == 'false'):
        chase.append(False)

    # climb = index 17
    if (row[17].lower() == 'true'):
        climb.append(True)
    elif (row[17].lower() == 'false'):
        climb.append(False)

    # eat = index 18
    if (row[18].lower() == 'true'):
        eat.append(True)
    elif (row[18].lower() == 'false'):
        eat.append(False)

    # forage = index 19
    if (row[19].lower() == 'true'):
        forage.append(True)
    elif (row[19].lower() == 'false'):
        forage.append(False)
```

```python
print("AM shift squirrel count: ", am_count)
print("PM shift squirrel count: ", pm_count)
```

```python
age_convert = []
for i in age:
  if (i == 'adult'):
    age_convert.append(1)
  elif (i == 'juvenile'):
    age_convert.append(0)
```
